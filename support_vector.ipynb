```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target 
df.head() 
```

```python
iris.target_names
```

```python
df['target_name'] = df['target'].apply(lambda x:iris.target_names[x])
df.head()
```

plotting scatter plot for analysis

```python
import matplotlib.pyplot as plt
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'])
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

From above plot we can conclude that any classificationn model will suit our dataset <br>
Here we are going to use support Vector Machine (SVM) for hyperlane classification using kernals

```python
# dependent and independent variables 

X = df.drop(columns=['target_name', 'target'])
y= df['target']
```

```python
# train test split 

X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

by using some argument we have tuned or model and score is 1.

```python
model_svc = SVC(C=10,gamma='scale', kernel='linear')
```

```python
model_svc.fit(X_train,y_train)
```

```python
model_svc.score(X_test, y_test)
```

```python
model_svc.predict([[7.1,3.0,5.9,2.1]])
```
